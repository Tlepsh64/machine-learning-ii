## I. Simplified two-step model for meme generation proccess

We are going to illustrate meme generation through generating points on a plane than a complex sentences, as relation between points is easier to visualise and interpret. However, the same procedure can be naturally extended to the original domain.

* **Step type choice.**
The parameter $\varrho$ determines what is the probability that we do abrupt change or a minor variation. The choices are independent throughout the evolution of the system. 

* **Minor variation.**
We randomly choose an existing meme $\boldsymbol{x}_i$ and modify it by adding white gaussian noise with low variance $\boldsymbol{\varepsilon}_i\sim\mathcal{N}(0, \sigma_1\boldsymbol{\mathrm I})$. The meme to be modified $\boldsymbol{x}_i$ is chosen uniformly form all the memes generated after the last seed generation.  

* **Seed generation.** 
We still choose an existing meme $\boldsymbol{x}_i$ and modify it by adding white gaussian noise with high variance $\boldsymbol{\varepsilon}_i\sim\mathcal{N}(0, \sigma_2\boldsymbol{\mathrm I})$. The meme to be modified $\boldsymbol{x}_i$ is chosen uniformly form all the memes generated after the last seed generation.



### Implementation

```python
def meme_generation(n: int, seed: List[float]=[0,0], rho: float=0.1, sigma1: float=0.01, sigma2: float=1)->DataFrame:
    z = rnd.uniform(size=n) <= rho
    x = np.full([n + 1,2], np.nan)
    x[0,:] = seed
    i0 = 0;
    source = np.full([n+1], -1, dtype=int)
    for i, seed in enumerate(z):
        source[i + 1] = rnd.choice(range(i0, i + 1))
        if seed:
            i0 = i + 1
            x[i + 1, :] = x[source[i + 1], :] + rnd.normal(scale=sigma2, size=2)
        else:
            x[i + 1, :] = x[source[i + 1], :] + rnd.normal(scale=sigma1, size=2)
            
    return DataFrame(x, columns = ['x', 'y']).assign(seed =  np.insert(z, 0, True)).assign(source = source)        
```

## II. Observed memes and the modification tree

In practice, the meme generation process is much more complicated and we do not know whether a particular meme $\boldsymbol{x}_i$ is a meme or not. We can model this by stating that we receive only randomly permuted set of memes $\boldsymbol{x}_1,\ldots,\boldsymbol{x}_n$ and our task is to reconstruct the meme generation process as best as we can.

For that note that each meme $\boldsymbol{x}_{i}$ is connected at least with one meme $\boldsymbol{x}_i$ through the following links:
* $\boldsymbol{x}_i$ was created from $\boldsymbol{x}_j$ by minor modification
* $\boldsymbol{x}_j$ was created from $\boldsymbol{x}_i$ by minor modification
* $\boldsymbol{x}_i$ was created from $\boldsymbol{x}_j$ by major modification
* $\boldsymbol{x}_j$ was created from $\boldsymbol{x}_i$ by major modification

Now if we omit the directionality information, then we must find a tree that covers all points $\boldsymbol{x}_i$ and edges have two colors: red for major changes and blue for minor changes. We refer to this graph as modification tree. See the figure below for the illustration.  

```python
df = meme_generation(100, rho=0.05, sigma1=0.01, sigma2=0.1)

sec = (df[1:]
       .assign(x0 = lambda df2: df.loc[df2['source'], 'x'].values)
       .assign(y0 = lambda df2: df.loc[df2['source'], 'y'].values)
       .rename(columns = {'x':'x1', 'y':'y1'})
       [['x0', 'y0', 'x1', 'y1', 'seed']]
) 

Delta = 0.5 * max(df['x'].max()-df['x'].min(),df['y'].max()-df['y'].min())
xc = 0.5 * (df['x'].max() + df['x'].min())
yc = 0.5 * (df['y'].max() + df['y'].min())

p = ggplot(df)
p = p + geom_point(aes(x = 'x', y = 'y', color = 'seed', alpha = 'seed'))
p = p + geom_segment(aes(x = 'x0', y =  'y0', xend = 'x1', yend = 'y1', color = 'seed'), data = sec, alpha = 0.3)
p = p + scale_alpha_manual(values=[0.3,1], breaks=[True, False], guide = False)
p = p + scale_color_manual(values=['blue','red'], breaks=[True, False], guide = False)
p = p + coord_fixed(ratio=1, xlim = [xc - Delta, xc + Delta], ylim = [yc - Delta, yc + Delta])
p = p + scale_x_continuous(name='') + scale_y_continuous(name='')
display(p)
```

## III. Maximum likelihood reconstruction of a modification tree in a cluster

Let us assume that all points are generated by nimor variations of previous points. Then we can ask what is the most likely generation tree $\mathcal{T}$. For fixed order of points $\boldsymbol{x}_0,\ldots,\boldsymbol{x}_n$ and source indexes $s_1,\ldots,s_n$, we can express 

\begin{align*}
\Pr[\mathcal{T}|\boldsymbol{x}_0,\ldots,\boldsymbol{x}_n,s_1,\ldots,s_n]=\prod_{i=1}^n \frac{1}{i-1}\cdot\Pr[\boldsymbol{x}_{s_i} \to\boldsymbol{x}_i]
\propto \Pr[\boldsymbol{x}_{s_i} \to\boldsymbol{x}_i]
\end{align*}

From this it is straightforward to observe that maximising source assignment $s_1,\ldots,s_n$ is such that 

\begin{align*}
  s_i = \mathrm{argmin}_{j<i} \Pr[\boldsymbol{x}_{j} \to\boldsymbol{x}_i]
\end{align*}

Thus we need to search only the right order of observations that maximises the probability $\Pr[\mathcal{T}|\boldsymbol{x}_0,\ldots,\boldsymbol{x}_n,s_1,\ldots,s_n]$. 
Note that log-likelihood can be expressed

\begin{align*}
\log \Pr[\mathcal{T}|\boldsymbol{x}_0,\ldots,\boldsymbol{x}_n,s_1,\ldots,s_n] = const + \log\Pr[\boldsymbol{x}_{s_i} \to\boldsymbol{x}_i]
\end{align*}

This probability is maximised if we find a spanning tree for a graph where each node pair $\boldsymbol{x}_i$ and $\boldsymbol{x}_j$ are connected with an edge of length $d_{ij}=-\log\Pr[\boldsymbol{x}_{s_i} \to\boldsymbol{x}_i]$. As $d_{ij}\geq 0$ we can use [Kruskal algorithm](https://en.wikipedia.org/wiki/Kruskal%27s_algorithm):

* Connect nodes with shortest edge provided that they are not already connected.
* Repeat the previous step until all nodes are connected. 

For white gaussian noise, the probability is inversely proportional to the distance square and thus the Kruskal algorithm is equivalent to the [single linkage clustering algorithm](https://en.wikipedia.org/wiki/Single-linkage_clustering): 

* [scipy.cluster.hierarchy.single](https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.single.html#scipy.cluster.hierarchy.single)
* [scipy.cluster.hierarchy.linkage](https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html#scipy.cluster.hierarchy.linkage)
* [sklearn.cluster.AgglomerativeClustering](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html#sklearn.cluster.AgglomerativeClustering). 

Unfortunately, the output of the implementations of single linkage clustering algorithm remembers only when a point is joined to a cluster and not what was the corresponding link $\boldsymbol{x}_i\leftrightarrow\boldsymbol{x}_j$. Hence, we implement Kruskals algorithm to get a better visualisation.   

### Barebone implementation of Kruskal algorithm

```python
def compute_distances(df: DataFrame) -> DataFrame:
    "Computes distances between row vectors of the dataframe"
    
    n = len(df)
    dist = np.full([int(0.5 * n *(n-1)), 3], np.nan)
    
    k = 0
    for i in range(len(df)):
        for j in range(i):
            d = np.sum((df.loc[i, ['x', 'y']] - df.loc[j, ['x', 'y']])**2)
            dist[k, :] = [i, j, d]
            k += 1
    
    return DataFrame(dist, columns = ['i', 'j', 'distance_sq']).astype({"i": int, "j": int}) 
```

```python
dist = compute_distances(df.loc[:5])
```

```python
def kruskal(dist: DataFrame) -> DataFrame:
    "Naive implementation of Kruskal algorithm"
    
    L = []
    clusters = []
    
    for _,row in dist.sort_values(['distance_sq']).iterrows():
    
        # Find clusters containing the end points of the edge  
        i0 = [i for i, cluster in enumerate(clusters) if row[0] in cluster]
        i1 = [i for i, cluster in enumerate(clusters) if row[1] in cluster]
        
        # Abort as we have a loop
        if i0 and (i0 == i1):
            continue

        L.append([int(row[0]), int(row[1])])

        # Update clusters
        if not i0 and not i1:
            clusters.append({row[0], row[1]})

        elif i0 and i1:
            clusters[i0[0]] |= clusters[i1[0]]
            del clusters[i1[0]]

        elif i0:
            clusters[i0[0]].add(row[1])

        elif i1:
            clusters[i1[0]].add(row[0])
    
    return DataFrame(L, columns = ['i', 'j'])
```

## IV. Maximum likelihood reconstruction for the entire graph

The inference task for the entire graph is analogous, except we must additionally consider whether the link $\boldsymbol{x}_{s_i} \to\boldsymbol{x}_i$ was generated by a minor modification or abrupt change. Let $c_1,\ldots,c_n$ be the indicator for the abrupt changes. Then we get

\begin{align*}
\Pr[\mathcal{T}|\boldsymbol{x}_0,\ldots,\boldsymbol{x}_n,s_1,\ldots,s_n, c_1,\ldots,c_n]=\prod_{i=1}^n \frac{1}{i-1}\cdot\Pr[\boldsymbol{x}_{s_i} \to\boldsymbol{x}_i|c_i]
\propto \Pr[\boldsymbol{x}_{s_i} \to\boldsymbol{x}_i|c_i]
\end{align*}

Again the order of observations $\boldsymbol{x}_0,\ldots,\boldsymbol{x}_n$ straightforwardly determines maximising values $s_1,\ldots,s_n$ and $c_1,\ldots,c_n$. Namely for $\boldsymbol{x_i}$ we must consider probabilities

\begin{align*}
\Pr[\boldsymbol{x}_j\to\boldsymbol{x_i}|c_i=0]\\
\Pr[\boldsymbol{x}_j\to\boldsymbol{x_i}|c_i=1]
\end{align*}

and choose the one with maximal probability. As before, we can define distances as log-likelihoods. However, there are now two edges between each node pair

\begin{align*}
d_{ij}^0=-\log \Pr[\boldsymbol{x}_j\to\boldsymbol{x_i}|c_i=0]\\
d_{ij}^1=-\log \Pr[\boldsymbol{x}_j\to\boldsymbol{x_i}|c_i=1]\\
\end{align*}

The latter is not an issue for Kruskal algorithm as we can keep only the shortest edge between two nodes.
